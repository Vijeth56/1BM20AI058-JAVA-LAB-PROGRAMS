```python
with open('data.txt', 'r') as file:
    edges = [line.strip().split() for line in file]
```

```python
import networkx as nx

G = nx.Graph()
G.add_edges_from(edges)
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
```

```python
import matplotlib.pyplot as plt
```

```python
# Betweenness Centrality
plt.figure()
plt.hist(list(betweenness_centrality.values()))
plt.title('Betweenness Centrality')
plt.xlabel('Centrality')
plt.ylabel('Frequency')

plt.show()
```

```python
# Degree Centrality
plt.figure()
plt.hist(list(degree_centrality.values()))
plt.title('Degree Centrality')
plt.xlabel('Centrality')
plt.ylabel('Frequency')

plt.show()
```

```python
# Closeness Centrality
plt.figure()
plt.hist(list(closeness_centrality.values()))
plt.title('Closeness Centrality')
plt.xlabel('Centrality')
plt.ylabel('Frequency')

plt.show()
```

```python
# Eigenvector Centrality
plt.figure()
plt.hist(list(eigenvector_centrality.values()))
plt.title('Eigenvector Centrality')
plt.xlabel('Centrality')
plt.ylabel('Frequency')
plt.show()
```
